2.  Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)?


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the discrete uniform distribution
low = 1      # Lower bound
high = 10    # Upper bound (inclusive)

# Generate random numbers from a discrete uniform distribution
random_numbers = np.random.randint(low, high + 1, size=1000)

# Calculate the frequency of each outcome
unique, counts = np.unique(random_numbers, return_counts=True)

# Calculate the PMF (probability mass function)
pmf = counts / len(random_numbers)

# Plot the PMF
plt.bar(unique, pmf, color='blue', alpha=0.7)
plt.title('Probability Mass Function (PMF) of a Discrete Uniform Distribution')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.xticks(unique)
plt.show()
```

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?


```python
def bernoulli_pdf(x, p):
    """
    Calculate the Probability Distribution Function (PDF) of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1).
    p (float): The probability of success (0 <= p <= 1).

    Returns:
    float: The probability of the outcome x.
    """
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        raise ValueError("x must be either 0 or 1")

# Example usage
p = 0.7  # Probability of success
x_values = [0, 1]

for x in x_values:
    print(f"P(X = {x}) = {bernoulli_pdf(x, p)}")
```

6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution?



```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the discrete uniform distribution
low = 1    # Lower bound
high = 10  # Upper bound

# Generate the outcomes (possible values in the distribution)
outcomes = np.arange(low, high + 1)

# Calculate the probability for each outcome
probability = 1 / (high - low + 1)

# Calculate the CDF
cdf = np.cumsum([probability] * len(outcomes))

# Plot the CDF
plt.step(outcomes, cdf, where='post', label='CDF', color='blue')
plt.title(f'Cumulative Distribution Function (CDF) of Discrete Uniform Distribution')
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')
plt.xticks(outcomes)
plt.grid(True)
plt.show()
```

9.  Write a Python function to calculate Z-scores from a dataset and plot them?


```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate the Z-scores of a dataset.

    Parameters:
    data (list or np.array): The dataset for which to calculate Z-scores.

    Returns:
    np.array: The Z-scores for the dataset.
    """
    mean = np.mean(data)         # Calculate the mean of the dataset
    std_dev = np.std(data)       # Calculate the standard deviation of the dataset
    z_scores = (data - mean) / std_dev  # Calculate Z-scores
    return z_scores

# Example dataset
data = np.random.normal(loc=0, scale=1, size=1000)  # Generate a dataset with a normal distribution

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plotting the Z-scores
plt.plot(z_scores, color='blue', label='Z-scores')
plt.axhline(0, color='red', linestyle='--', label='Mean (Z = 0)')
plt.title('Z-scores of the Dataset')
plt.xlabel('Index')
plt.ylabel('Z-score')
plt.legend()
plt.show()
```

10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.


```python
import numpy as np
import matplotlib.pyplot as plt

def central_limit_theorem(non_normal_data, sample_size, num_samples):
    """
    Implement the Central Limit Theorem (CLT).

    Parameters:
    non_normal_data (np.array): The non-normal data from which to sample.
    sample_size (int): The size of each sample.
    num_samples (int): The number of samples to draw.

    Returns:
    np.array: The means of the sampled data.
    """
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(non_normal_data, size=sample_size, replace=True)
        sample_means.append(np.mean(sample))
    return np.array(sample_means)

# Generate a non-normal distribution (e.g., Uniform Distribution)
low = 0
high = 10
size = 100000  # Size of the non-normal data
non_normal_data = np.random.uniform(low, high, size)

# Parameters for the CLT simulation
sample_size = 50  # Sample size for each iteration
num_samples = 1000  # Number of samples to take

# Apply the Central Limit Theorem
sample_means = central_limit_theorem(non_normal_data, sample_size, num_samples)

# Plot the distribution of the sample means
plt.hist(sample_means, bins=30, edgecolor='black', density=True, alpha=0.7, color='blue')

# Plot the normal distribution for comparison (Theoretical)
from scipy.stats import norm
mu = np.mean(sample_means)
std_dev = np.std(sample_means)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std_dev)
plt.plot(x, p, 'k', linewidth=2, label='Normal Approximation')

# Add title and labels
plt.title(f'Central Limit Theorem: Distribution of Sample Means\n'
          f'Non-Normal Distribution (Uniform) Sample Size={sample_size}, Num Samples={num_samples}')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def central_limit_theorem_normal(mean, std_dev, sample_size, num_samples):
    """
    Simulate the Central Limit Theorem (CLT) for a normal distribution.

    Parameters:
    mean (float): The mean of the normal distribution.
    std_dev (float): The standard deviation of the normal distribution.
    sample_size (int): The size of each sample.
    num_samples (int): The number of samples to take.

    Returns:
    np.array: The means of the sampled data.
    """
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.normal(mean, std_dev, sample_size)  # Draw random sample
        sample_means.append(np.mean(sample))  # Calculate and store sample mean
    return np.array(sample_means)

# Parameters for the normal distribution
mean = 0      # Mean of the normal distribution
std_dev = 1   # Standard deviation of the normal distribution
sample_size = 50  # Size of each sample
num_samples = 1000  # Number of samples to draw

# Apply the Central Limit Theorem
sample_means = central_limit_theorem_normal(mean, std_dev, sample_size, num_samples)

# Plot the distribution of the sample means
plt.hist(sample_means, bins=30, edgecolor='black', density=True, alpha=0.7, color='blue')

# Plot the original normal distribution for comparison (Theoretical)
from scipy.stats import norm
mu = np.mean(sample_means)
std_dev_sample_means = np.std(sample_means)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std_dev_sample_means)
plt.plot(x, p, 'k', linewidth=2, label='Normal Approximation')

# Add title and labels
plt.title(f'Central Limit Theorem: Distribution of Sample Means\n'
          f'Normal Distribution Sample Size={sample_size}, Num Samples={num_samples}')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def generate_binomial_data(n, p, size):
    """
    Generate random variables from a binomial distribution and calculate their corresponding probabilities.

    Parameters:
    n (int): Number of trials.
    p (float): Probability of success on each trial.
    size (int): Number of random samples to generate.

    Returns:
    np.array: Random variables (outcomes) from the binomial distribution.
    np.array: Corresponding probabilities for each outcome.
    """
    # Generate random variables (outcomes) from the binomial distribution
    outcomes = np.random.binomial(n, p, size)

    # Calculate probabilities using the PMF of the binomial distribution
    probabilities = binom.pmf(outcomes, n, p)

    return outcomes, probabilities

# Parameters for the binomial distribution
n = 10       # Number of trials
p = 0.5      # Probability of success
size = 1000  # Number of random samples

# Generate random variables and corresponding probabilities
outcomes, probabilities = generate_binomial_data(n, p, size)

# Plot the histogram of outcomes
plt.hist(outcomes, bins=range(n+2), density=True, alpha=0.6, color='blue', label='Sampled Outcomes')

# Plot the theoretical binomial distribution
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)
plt.plot(x, pmf, 'ro', label='Theoretical Binomial PMF')

# Add title and labels
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.legend()

# Show the plot
plt.show()
```

18. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal?


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data, x):
    """
    Calculate the Z-score for a given data point x.

    Parameters:
    data (np.array): The dataset (list or array of numbers).
    x (float): The data point for which to calculate the Z-score.

    Returns:
    float: The Z-score for the data point x.
    """
    mean = np.mean(data)  # Calculate the mean of the dataset
    std_dev = np.std(data)  # Calculate the standard deviation of the dataset
    z_score = (x - mean) / std_dev  # Calculate the Z-score
    return z_score

def plot_standard_normal_distribution():
    """
    Plot the standard normal distribution curve.
    """
    # Generate values from -4 to 4 (since most of the distribution is within this range)
    x = np.linspace(-4, 4, 1000)

    # Calculate the PDF of the standard normal distribution
    pdf = norm.pdf(x, 0, 1)

    # Plot the standard normal distribution
    plt.plot(x, pdf, label="Standard Normal Distribution", color='blue')
    plt.title("Standard Normal Distribution (Mean = 0, Std = 1)")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")

    # Add grid and legend
    plt.grid(True)
    plt.legend()
    plt.show()

# Example data
data = np.random.normal(10, 2, 1000)  # Generate data from a normal distribution (mean=10, std=2)

# Given data point to calculate Z-score for
x = 12  # Example data point

# Calculate the Z-score for the data point
z_score = calculate_z_score(data, x)

# Print the Z-score
print(f"The Z-score for the data point {x} is: {z_score}")

# Plot the standard normal distribution
plot_standard_normal_distribution()

# Compare the Z-score on the standard normal distribution
plt.axvline(x=z_score, color='red', linestyle='--', label=f'Z-score for {x} = {z_score:.2f}')
plt.legend()
plt.show()
```

19.  Implement hypothesis testing using Z-statistics for a sample dataset?


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform Z-test for hypothesis testing.

    Parameters:
    sample_data (np.array): The sample data (list or array of numbers).
    population_mean (float): The population mean under the null hypothesis.
    population_std (float): The population standard deviation (assumed known).
    alpha (float): Significance level (default is 0.05).

    Returns:
    tuple: Z-statistic, p-value, and whether to reject the null hypothesis.
    """
    # Sample mean and sample size
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Calculate Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))  # Two-tailed test

    # Determine if we reject the null hypothesis
    reject_null = p_value < alpha

    return z_statistic, p_value, reject_null

# Sample data
np.random.seed(42)
sample_data = np.random.normal(105, 15, 50)  # Sample data (e.g., 50 data points from normal distribution)
population_mean = 100  # Hypothesized population mean
population_std = 15    # Assumed population standard deviation

# Perform Z-test
z_statistic, p_value, reject_null = z_test(sample_data, population_mean, population_std, alpha=0.05)

# Print the results
print(f"Z-statistic: {z_statistic:.3f}")
print(f"P-value: {p_value:.3f}")
if reject_null:
    print("Reject the null hypothesis (the sample mean is significantly different from the population mean).")
else:
    print("Fail to reject the null hypothesis (there is no significant difference).")

# Plotting the standard normal distribution and the Z-statistic
x = np.linspace(-4, 4, 1000)
pdf = stats.norm.pdf(x, 0, 1)

plt.plot(x, pdf, label="Standard Normal Distribution", color='blue')

# Plot the Z-statistic
plt.axvline(x=z_statistic, color='red', linestyle='--', label=f'Z-statistic = {z_statistic:.2f}')
plt.fill_between(x, pdf, where=(x >= z_statistic), color='red', alpha=0.3)

plt.title(f"Z-Test for Hypothesis Testing\nZ-statistic = {z_statistic:.2f}")
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for the mean of the sample data.

    Parameters:
    data (np.array): The sample data (list or array of numbers).
    confidence_level (float): The confidence level (default is 95%).

    Returns:
    tuple: The lower and upper bounds of the confidence interval.
    """
    # Sample size
    n = len(data)

    # Sample mean and sample standard deviation
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Using ddof=1 for sample standard deviation

    # Degrees of freedom
    df = n - 1

    # t-critical value for the given confidence level
    t_critical = stats.t.ppf((1 + confidence_level) / 2, df)

    # Margin of error
    margin_of_error = t_critical * (sample_std / np.sqrt(n))

    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example data (e.g., sample of exam scores)
np.random.seed(42)
data = np.random.normal(75, 10, 30)  # Mean of 75, standard deviation of 10, and 30 data points

# Calculate the confidence interval
confidence_level = 0.95
lower, upper = calculate_confidence_interval(data, confidence_level)

# Print the result
print(f"Confidence Interval ({confidence_level*100}%): [{lower:.2f}, {upper:.2f}]")

```

```python
import numpy as np
import scipy.stats as stats

# Step 1: Generate data from a normal distribution
np.random.seed(0)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=100)  # Mean = 50, Std Dev = 10, n = 100

# Step 2: Calculate the sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)
sample_size = len(data)
standard_error = sample_std / np.sqrt(sample_size)

# Step 3: Calculate the confidence interval for the mean
confidence_level = 0.95
z_value = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)  # t-distribution for 95% CI
margin_of_error = z_value * standard_error

# Confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

sample_mean, ci_lower, ci_upper  # Output the results: mean and confidence interval bounds
```

27. Implement Z-score normalization on a dataset and visualize its transformation?


```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate data from a normal distribution
np.random.seed(0)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean = 50, Std Dev = 10, n = 1000

# Step 2: Calculate the mean and standard deviation of the original data
mean = np.mean(data)
std_dev = np.std(data)

# Step 3: Apply Z-score normalization
data_normalized = (data - mean) / std_dev

# Step 4: Plot original and normalized data
plt.figure(figsize=(12, 6))

# Plot original data
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Data (Before Normalization)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot normalized data
plt.subplot(1, 2, 2)
plt.hist(data_normalized, bins=30, color='lightgreen', edgecolor='black')
plt.title('Normalized Data (After Z-score Normalization)')
plt.xlabel('Z-score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

28.  Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.


```python
import numpy as np
import scipy.stats as stats

# Function to calculate skewness and kurtosis
def calculate_skewness_kurtosis(data):
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data)
    return skewness, kurtosis

# Generate a dataset from a normal distribution
mean = 0
std_dev = 1
size = 1000
data = np.random.normal(loc=mean, scale=std_dev, size=size)

# Calculate skewness and kurtosis
skewness, kurtosis = calculate_skewness_kurtosis(data)

# Print the results
print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis: {kurtosis:.4f}")
```
